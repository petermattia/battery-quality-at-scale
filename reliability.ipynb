# Reliability

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import rcParams
```

## Code

Define figure dimensions:

```python
fig_width = 3.25 # single column width
fig_width_2col = 2 * fig_width # double column width
fig_height = (3 / 4) * fig_width
```

Define default plot settings:

```python
rcParams['figure.autolayout'] = True
rcParams['lines.linewidth'] = 1.0
rcParams['font.size'] = 8
rcParams['font.sans-serif'] = 'Arial'
rcParams['mathtext.fontset'] = 'custom'
rcParams['mathtext.rm'] = 'Arial'
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

Do some basic math:

```python
# Generate the grid
x = np.linspace(1, 5000, 500) # cells in pack
y = np.linspace(0.999, 1, 500) # cell-level reliability
X, Y = np.meshgrid(x, y)

# Compute pack-level reliability
C = Y**X
```

Plot:

```python
plt.figure(figsize=(fig_width, fig_height))
plt.pcolormesh(X, Y, C, shading='auto')
plt.colorbar(label='Pack-level reliability, $P_{pack} = P_{cell}^{n}$')
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('Number of cells in the pack, $n$')
plt.ylabel('Cell-level reliability, $P_{cell}$')
plt.show()

plt.savefig("figures/reliability.pdf")
```
