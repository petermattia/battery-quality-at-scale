# Reliability

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib import rcParams
```

## Code

Define figure dimensions:

```python
fig_width_mm = 88 # single column width for Nature Comms

mm = 1/25.4  # millimeters in inches
fig_width_inches = fig_width_mm * mm

fig_height_inches = (3 / 4) * fig_width_inches

print(fig_width_inches, fig_height_inches)
```

Define default plot settings:

```python
rcParams['figure.autolayout'] = True
rcParams['lines.linewidth'] = 1.0
rcParams['font.size'] = 8
rcParams['font.sans-serif'] = 'Arial'
rcParams['mathtext.fontset'] = 'custom'
rcParams['mathtext.rm'] = 'Arial'
rcParams['pdf.fonttype'] = 42
rcParams['ps.fonttype'] = 42
```

Do some basic math:

```python
# Generate the grid
x = np.linspace(1, 5000, 250) # cells in pack
y = np.linspace(0.9999, 1, 250) # cell-level reliability
X, Y = np.meshgrid(x, y)

# Compute pack-level reliability
C = Y**X
```

Define a custom formatter function for fixed point notation:

```python
def format_ticks(value, pos):
    # Format the value to display with 5 decimal places
    return '{:.5f}'.format(value)
```

Plot:

```python
plt.figure(figsize=(fig_width_inches, fig_height_inches))
plt.imshow(C, aspect='auto', origin='lower', extent=[-0.1, 5000, 0.9999, 1], cmap="viridis_r")

plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))

plt.colorbar(label='Pack-level reliability, $R_{pack} = R_{cell}^{n}$')
plt.xlabel('Number of cells in the pack, $n$')
plt.ylabel('Cell-level reliability, $R_{cell}$')

plt.savefig("reliability.pdf", dpi=300, bbox_inches="tight", format="pdf")
```
